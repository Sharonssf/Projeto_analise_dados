````python
#_= plt.hist(rh.Age)
#_= plt.title("Gráfico de idade por funcionários")


#Não sei qual ficou melhor para entender, para mim ambos estão meio confusos

_ = plt.hist(rh.Age, bins=20, edgecolor='g')
_ = plt.xlabel("Idade")
_ = plt.ylabel("Contagem de Funcionários")
_ = plt.title("Distribuição de Idades dos Funcionários")
````

````python
#_= plt.hist(rh.MonthlyIncome)
#_= plt.title("Gráfico de salário por funcionários")

#Acho q assim fica melhor q o outro para entender
_ = plt.hist(rh.MonthlyIncome, bins=20, edgecolor='g')
_ = plt.xlabel("Salário Mensal")
_ = plt.ylabel("Contagem de Funcionários")
_ = plt.title("Distribuição de Salários Mensais")
````

````python
#Vendo a quantidade de generos presentes na empresa
rh.Gender.value_counts()
````

````python
_ = rh.EducationField.value_counts().plot.pie(autopct='%1.1f%%')
_ = plt.title("Contagem por campo educacional")
````

````python
#Não botei o _=, justamente para mostrar q se não botar fica essa frase nada a ver ai
sns.boxplot(x = "Gender", y = "MonthlyIncome", data = rh)
````

````python
#Não entendi mt bem esse grafico então fiz um outro q mostra a mesma coisa mas mais facil de entender
_= sns.scatterplot(x="YearsAtCompany", y = "MonthlyIncome", data = rh)
````

````python
# Calculando a média de salário mensal por anos na empresa
average_income_by_years = rh.groupby("YearsAtCompany")["MonthlyIncome"].mean().reset_index()

# Criando o gráfico de barras
_ = sns.barplot(x="YearsAtCompany", y="MonthlyIncome", data=average_income_by_years)
_ = plt.xlabel("Anos na Empresa")
_ = plt.ylabel("Média de Salário Mensal")
_ = plt.title("Média de Salário Mensal por Anos na Empresa")

#ficou bonitinho
````
